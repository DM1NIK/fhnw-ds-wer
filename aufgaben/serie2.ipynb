```python
# a)
p_v = 1 * 5/6 # Probability of getting two different numbers

div, counter = 0, 0
for i in range(1, 6 + 1):
    for j in range(1, 6 + 1):
        if j > i:
            counter += 1
            div += (1 if j % i == 0 else 0)

print(div, counter, div/counter)

print("Probability: ", p_v * (div/counter), 4/9)
```

```python
# b)
matching, counter = 0, 0
for i in range(1, 6 + 1):
    for j in range(1, 6 + 1):
        counter += 1
        if abs(j - i) == 2:
            matching += 1

print(matching, counter, matching/counter, 2/9)
```

```python
# c)
matching, counter = 0, 0
for i in range(1, 6 + 1):
    for j in range(1, 6 + 1):
        counter += 1
        if abs(j - i) >= 2:
            matching += 1

print(matching, counter, matching/counter, 5/9)
```

```python
# d)
matching, counter = 0, 0
for i in range(1, 6 + 1):
    for j in range(1, 6 + 1):
        counter += 1
        if j * 2 == i or j == i * 2:
            matching += 1

print(matching, counter, matching/counter, 1/6)
```

```python
# a)
matching, counter = 0, 0
for i in range(1, 2500 + 1):
    counter += 1
    if str(i)[0] == '1':
        matching += 1

print(matching, counter, matching/counter, 1111/2500)
```

```python
data = [60, 52, 61, 60, 54, 58]
mean, median = np.mean(data), np.median(data)

# a)
matching, counter = 0, len(data)
for e in data:
    if e < mean:
        matching += 1

print(matching, counter, matching/counter, 1/3)
```

```python
# b)
matching, counter = 0, len(data)
for e in data:
    if e < median:
        matching += 1

print(matching, counter, matching/counter, 1/2)
```

```python
event_space = []
# All possible combinations of samples of size 4 with two different values "0" and "1"
for i in range(2**4):
    event_space.append([int(x) for x in bin(i)[2:].zfill(4)])

print(event_space)

# a)
matching, counter = 0, len(event_space)
for e in event_space:
    if e[0] == 0:
        matching += 1

A = matching/counter
print(matching, counter, matching/counter, 8/16)
```

```python
# b)
matching, counter = 0, len(event_space)
for e in event_space:
    if e[0] == 0 and sum(e[1:]) == 3:
        matching += 1

B = matching/counter
print(matching, counter, matching/counter, 1/16)
```

```python
# A and C
matching, counter = 0, len(event_space)
for e in event_space:
    if e[0] == 0 or sum(e[1:]) == 2:
        matching += 1
print(matching, counter, matching/counter, 11/16)
```
