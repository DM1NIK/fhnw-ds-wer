```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import util
```

```python
# Comapring ttest function to scipy ttest_1samp function

sample = [33,31,30,29,24,27,28,31,25,23]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1) # ddof=1 for sample std (n-1)

print("Sample mean: ", sample_mean)
print("Sample std: ", sample_std)

alpha = .05

util.calc_ttest(n=len(sample), u=sample_mean, u0=27, std0=sample_std, significance=alpha, side="right", visualize=True) # set std0 to the sample std to get the same result as scipy ttest_1samp
util.calc_ttest_scipy(sample=sample, u0=27, significance=alpha, alternative='greater')
```
